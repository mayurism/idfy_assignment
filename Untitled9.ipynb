## Data preprocessing

Now we will define a function
args- pdf path
output -extracted text from document
as documents are lengthy and we just to need to classfiy them by their form numbers, as all form numbers are on the 
first page, we will just convert first page of dcoument into text

```python
def extract_text_from_scanned_pdf(pdf_path):
    # Open the PDF file
        pdf_document = fitz.open(pdf_path)

    # Iterate through each page
#     for page_number in range(pdf_document.page_count):
        # Extract the image from the page
        page = pdf_document[0]
        image_list = page.get_images(full=True)
#         print(image_list)

        # Iterate through each image on the page
        for image_index, img in enumerate(image_list):
#             print(image_index)
            base_image = pdf_document.extract_image(img[0])
            image_bytes = base_image["image"]

            # Save the image to a temporary file
            image_path = f"temp_image_{0}_{image_index}.png"
            with open(image_path, "wb") as img_file:
                img_file.write(image_bytes)

            # Use Tesseract OCR to extract text from the image
            extracted_text = pytesseract.image_to_string(Image.open(image_path), lang='eng')

            # Print or use the extracted text as needed
#             print(f"Page {0 + 1}, Image {image_index + 1}:\n{extracted_text}\n")
            return extracted_text

    # Close the PDF document
        pdf_document.close()
```

## cleaning data

in preprocessing we will first convert text into lowercase, then remove special characters, we will not remove numbers 
and '-'as they are present in most of the form number, then we will replace multiple newlines with single newline

```python
def clean_text(x):
    x = x.lower() #converting whole text into lowercase
    print(x)
    x = re.sub(r'[^a-zA-Z0-9\n\s-]', '', x) #replacing all characters that are not alphanumeric or '-'
    print('------------------------')
    print(x)
    x = re.sub(r'\n+', '\n', x)  #replacing multiple newlines with single newline
    print('--------------------------')
    print(x)

    return x

    
```

now we have done basic preprocessing, to build a classifier we will need labeled data, now as we dont have labeled data,
we will manually extract form number regex pattern and save it in new column as 'label'.
then we will vectorize the data and build model

```python
def find_label(x):
    pattern = re.compile(r'\bform\s+[a-z\d_-]+\b', re.MULTILINE)
    matches = pattern.findall(x)
#             print(matches[0])
    if matches:
            return matches[0]
    else:
            return('other')
```

now we will encode label column using label encoder as models can not process alphabets

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['label'])
```

```python
df.head(2)
```

```python
# splitting the data into train and test for model training 
x_train, x_test, y_train, y_test = train_test_split(df['clean_text'], df['class'], test_size=0.2, random_state=1, stratify=df['class'])
```

## lets initialize a tfidf vectorizer to convert clean_text into vectors, in this first I am going to create TFIDF vectors whole scanned data

```python
tf_vector  = TfidfVectorizer(ngram_range=(1,2))
x_train_vec = tf_vector.fit_transform(x_train)
x_train_vec.shape
```

```python
x_test_vec = tf_vector.transform(x_test)
x_test_vec.shape
```

```python
lg_model = LogisticRegression(class_weight='balanced') #defining logistic regression 
lg_model.fit(x_train_vec, y_train)
y_pred = lg_model.predict(x_test_vec)
print('Accuracy score: ',accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n',confusion_matrix(y_test, y_pred))
print('Classification Report:\n',classification_report(y_test, y_pred))
```

if we observe above results, we are getting good accuracy of 93% and precision is also better

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()# using random forest for alternate approach
rf.fit(x_train_vec, y_train)
y_pred_rf = rf.predict(x_test_vec)
print(accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
```

As we can see, accuracy and weighted avg precision is better with random forest

this is another approach i was trying using Kmeans clustering, but we get elbow at 9 clusters, so if we do not want to label the data and classify the documents, we can go with this approach, and for form numbers we will again have to extract with regex pattern.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
```

```python
tftdf_matrix = tf_vector.fit_transform(df['clean_text_small'])
svd = TruncatedSVD(n_components=2, random_state=42)
tfidf_matrix_svd = svd.fit_transform(tftdf_matrix)
```

```python
distortions = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(tftdf_matrix)
    distortions.append(kmeans.inertia_)

# Plot the elbow method graph
plt.plot(range(1, 11), distortions, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion (Within-Cluster Sum of Squares)')
plt.show()
```

```python
optimal_k = 9  # Choose the number based on the elbow method or other criteria
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(tftdf_matrix)

# Get cluster assignments
labels = kmeans.labels_

# Visualize clusters using dimensionality-reduced data
plt.scatter(tfidf_matrix_svd[:, 0], tfidf_matrix_svd[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8, edgecolors='w')
plt.title('K-Means Clustering of TF-IDF Vectors')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```
